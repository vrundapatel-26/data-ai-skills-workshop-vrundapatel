```python
def _safe_int_conversion(value):
    try:
        return int(value.strip())
    except (ValueError, AttributeError):
        return None


def _safe_float_conversion(value):
    try:
        return float(value.strip())
    except (ValueError, AttributeError):
        return None


def _convert_date_format(date_str):
    """Convert YYYY/MM/DD to YYYY-MM-DD for BigQuery"""
    if date_str and date_str.strip():
        return date_str.strip().replace('/', '-')
    return None


def parse_message(message):
    fields = [f.strip() for f in message.strip().split(',')]

    # Handle potential missing fields by providing default values
    mt = fields[0] if len(fields) > 0 else None
    tt = _safe_int_conversion(fields[1]) if len(fields) > 1 else None
    sid = fields[2] if len(fields) > 2 else None
    aid = fields[3] if len(fields) > 3 else None
    hex_code = fields[4] if len(fields) > 4 else None
    fid = fields[5] if len(fields) > 5 else None
    dmg = _convert_date_format(fields[6]) if len(fields) > 6 else None
    tmg = fields[7] if len(fields) > 7 else None
    dml = _convert_date_format(fields[8]) if len(fields) > 8 else None
    tml = fields[9] if len(fields) > 9 else None
    cs = fields[10] if len(fields) > 10 else None
    alt = _safe_int_conversion(fields[11]) if len(fields) > 11 else None
    gs = _safe_int_conversion(fields[12]) if len(fields) > 12 else None
    trk = _safe_int_conversion(fields[13]) if len(fields) > 13 else None
    lat = _safe_float_conversion(fields[14]) if len(fields) > 14 else None
    lng = _safe_float_conversion(fields[15]) if len(fields) > 15 else None
    vr = _safe_int_conversion(fields[16]) if len(fields) > 16 else None
    sq = fields[17] if len(fields) > 17 else None
    alrt = _safe_int_conversion(fields[18]) if len(fields) > 18 else None
    emer = _safe_int_conversion(fields[19]) if len(fields) > 19 else None
    spi = _safe_int_conversion(fields[20]) if len(fields) > 20 else None
    gnd = _safe_int_conversion(fields[21]) if len(fields) > 21 else None

    return {
        "MT": mt,
        "TT": tt,
        "SID": sid,
        "AID": aid,
        "Hex": hex_code,
        "FID": fid,
        "DMG": dmg,
        "TMG": tmg,
        "DML": dml,
        "TML": tml,
        "CS": cs,
        "Alt": alt,
        "GS": gs,
        "Trk": trk,
        "Lat": lat,
        "Lng": lng,
        "VR": vr,
        "Sq": sq,
        "Alrt": alrt,
        "Emer": emer,
        "SPI": spi,
        "Gnd": gnd
    }
```

```python
from google.cloud.pubsub_v1 import SubscriberClient

subscription_id = 'flight-transponder-sub1'
project_id = 'qwiklabs-gcp-04-14e3a67e50ad'

subscriber = SubscriberClient()
subscription_path = subscriber.subscription_path(project_id, subscription_id)

def callback(message):
    try:
        data = message.data.decode('utf-8')
        parsed_data = parse_message(data)

        # Insert into BigQuery
        table_id = 'transponder_data'
        table_ref = client.dataset(dataset_id).table(table_id)
        table = client.get_table(table_ref)

        rows_to_insert = [parsed_data]
        errors = client.insert_rows_json(table, rows_to_insert)

        if errors:
            print(f"Errors inserting row: {errors}")
        else:
            print("Successfully inserted row.")

        message.ack()

    except Exception as e:
        print(f"Error processing message: {e}")
        message.nack()

streaming_pull_future = subscriber.subscribe(subscription_path, callback=callback)

print(f"Listening for messages on {subscription_path}...")

# Keep the main thread alive to receive messages
try:
    streaming_pull_future.result()
except KeyboardInterrupt:
    streaming_pull_future.cancel()
    streaming_pull_future.result()
```
